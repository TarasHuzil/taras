```python
from random import randint

for n in range(30):
    x = randint(-10, 10)
    try:
        y = 7/x
    except ZeroDivisionError as e:
        # Це так званий обробник помилок
        y = f"Виникла помилка: {e}"

    print(f"{n+1}) При діленні 7 на {x} буде: {y}")
```
